```python
from tkinter import *
from tkinter import messagebox as mb
import json
import pyttsx3
import matplotlib.pyplot as plt

# Creating a GUI Window
guiWindow = Tk()
guiWindow.geometry("1000x450")
guiWindow.title("FinQuest")

class myQuiz:
    def __init__(self):
        self.quesNumber = 0
        self.optSelected = IntVar(value=0)
        self.engine = pyttsx3.init()
        self.difficulty = StringVar()
        self.difficulty.set("easy")  # Default difficulty
        self.displayTitle()
        self.displayDifficultySelection()
        self.start_button = Button(guiWindow, text="Start Quiz", command=self.startQuiz)
        self.start_button.place(x=450, y=300)

    def displayDifficultySelection(self):
        Label(guiWindow, text="Select Difficulty:", font=("ariel", 16)).place(x=400, y=100)
        Radiobutton(guiWindow, text="Easy", variable=self.difficulty, value="easy", font=("ariel", 14)).place(x=400, y=150)
        Radiobutton(guiWindow, text="Medium", variable=self.difficulty, value="medium", font=("ariel", 14)).place(x=400, y=200)
        Radiobutton(guiWindow, text="Hard", variable=self.difficulty, value="hard", font=("ariel", 14)).place(x=400, y=250)

    def startQuiz(self):
        self.loadQuestions()
        self.start_button.destroy()
        for widget in guiWindow.winfo_children():
            if isinstance(widget, Radiobutton):
                widget.destroy()
        self.time_remaining = 60
        self.timer_label = Label(guiWindow, text=f"Time Left: {self.time_remaining} sec", font=("cambria", 16))
        self.timer_label.place(x=700, y=80)
        self.displayQuestion()
        self.options = self.radioButtons()
        self.displayOptions()
        self.buttons()
        self.dataSize = len(self.question)
        self.rightAnswer = 0
        self.update_timer()

    def loadQuestions(self):
        difficulty = self.difficulty.get()
        self.question = data[difficulty]['ques']
        self.opts = data[difficulty]['choices']
        self.answer = data[difficulty]['ans']
        self.feedback = data[difficulty]['feedback']

    def update_timer(self):
        if self.time_remaining > 0:
            self.timer_label.config(text=f"Time Left: {self.time_remaining} sec")
            self.time_remaining -= 1
            guiWindow.after(1000, self.update_timer)
        else:
            mb.showinfo("Time's up!", "Time is over!")
            self.displayResult()
            guiWindow.destroy()

    def speak(self, text):
        self.engine.say(text)
        self.engine.runAndWait()

    def visualize_performance(self):
        correct = self.rightAnswer
        wrong = self.dataSize - correct
        categories = ['Correct', 'Wrong']
        counts = [correct, wrong]

        plt.figure(figsize=(8, 6))
        plt.bar(categories, counts, color=['green', 'red'])
        plt.title('Quiz Performance')
        plt.ylabel('Number of Answers')
        plt.ylim(0, self.dataSize)
        plt.show()

    def displayResult(self):
        wrongCount = self.dataSize - self.rightAnswer
        rightAnswer = f"Correct: {self.rightAnswer}"
        wrongAnswer = f"Wrong: {wrongCount}"
        the_score = int(self.rightAnswer / self.dataSize * 100) if self.dataSize > 0 else 0
        the_result = f"Score: {the_score}%"
        mb.showinfo("Result", f"{the_result} \n{rightAnswer} \n{wrongAnswer}")
        self.speak(f"You scored {the_score} percent.")
        self.visualize_performance()

    def checkAnswer(self, quesNumber):
        return self.optSelected.get() == self.answer[quesNumber]

    def nextButton(self):
        selected_option = self.optSelected.get() - 1
        if selected_option < 0:
            mb.showwarning("No Selection", "Please select an option before proceeding.")
            return
       
        is_correct = self.checkAnswer(self.quesNumber)
        feedback_message = self.feedback[self.quesNumber][selected_option]
       
        if is_correct:
            self.rightAnswer += 1
            feedback_text = f"Correct! {feedback_message}"
            self.speak("Correct answer!")
        else:
            correct_option = self.opts[self.quesNumber][self.answer[self.quesNumber] - 1]
            feedback_text = f"Wrong! Correct answer: {correct_option}\n{feedback_message}"
            self.speak(f"Wrong! The correct answer is {correct_option}.")
        
        guiWindow.after(2000, lambda: mb.showinfo("Feedback", feedback_text))
        guiWindow.update()
        self.disableButtons()
        guiWindow.after(2000, self.transitionToNext)

    def buttons(self):
        next_button = Button(
            guiWindow,
            text="Next",
            command=self.nextButton,
            width=10,
            bg="blue",
            fg="white",
            font=("ariel", 16, "bold")
        )
        next_button.place(x=350, y=380)

        quit_button = Button(
            guiWindow,
            text="Quit",
            command=guiWindow.destroy,
            width=5,
            bg="black",
            fg="white",
            font=("ariel", 16, "bold")
        )
        quit_button.place(x=700, y=50)

        read_question_button = Button(
            guiWindow,
            text="Read Question",
            command=self.readQuestion,
            width=15,
            bg="orange",
            fg="black",
            font=("ariel", 16, "bold")
        )
        read_question_button.place(x=100, y=320)

        read_options_button = Button(
            guiWindow,
            text="Read Options",
            command=self.readOptions,
            width=15,
            bg="purple",
            fg="white",
            font=("ariel", 16, "bold")
        )
        read_options_button.place(x=600, y=320)

    def readQuestion(self):
        self.speak(self.question[self.quesNumber])

    def readOptions(self):
        options_text = " ".join(self.opts[self.quesNumber])
        self.speak(f"The options are: {options_text}")

    def displayOptions(self):
        val = 0
        self.optSelected.set(0)
        for opt in self.opts[self.quesNumber]:
            self.options[val]['text'] = opt
            val += 1

    def displayQuestion(self):
        quesNumber = Label(
            guiWindow,
            text=self.question[self.quesNumber],
            width=60,
            font=('ariel', 16, 'bold'),
            anchor='w'
        )
        quesNumber.place(x=70, y=100)

    def displayTitle(self):
        myTitle = Label(
            guiWindow,
            text="FinQuest",
            width=50,
            bg="red",
            fg="white",
            font=("ariel", 20, "bold")
        )
        myTitle.place(x=0, y=2)

    def radioButtons(self):
        qList = []
        y_pos = 150
        while len(qList) < 4:
            radio_button = Radiobutton(
                guiWindow,
                text=" ",
                variable=self.optSelected,
                value=len(qList) + 1,
                font=("ariel", 14)
            )
            qList.append(radio_button)
            radio_button.place(x=100, y=y_pos)
            y_pos += 40
        return qList

    def transitionToNext(self):
        self.quesNumber += 1
        if self.quesNumber == self.dataSize:
            self.displayResult()
            guiWindow.destroy()
        else:
            self.displayQuestion()
            self.displayOptions()
            self.enableButtons()

    def disableButtons(self):
        for btn in self.options:
            btn.config(state=DISABLED)

    def enableButtons(self):
        for btn in self.options:
            btn.config(state=NORMAL)

# Load quiz data from JSON file
with open('QS6.json') as json_file:
    data = json.load(json_file)

# Create quiz instance and start
quiz = myQuiz()
guiWindow.mainloop()
```
